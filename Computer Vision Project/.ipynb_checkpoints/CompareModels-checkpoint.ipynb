# Comparing 3 Transfer Learning Models

```python
# Importing necessary modules
import os
import numpy as np
import keras
from keras import layers
from tensorflow import data as tf_data
import matplotlib.pyplot as plt
```

```python
# Generating testing dataset for comparison of ROC curves
image_size = (224, 224)

batch_size = 10

test_ds = keras.utils.image_dataset_from_directory(
    "test",
    seed=1337,
    image_size=image_size,
    batch_size=batch_size,
)
```

```python
# Visualizing testing images
plt.figure(figsize=(10, 10))
for images, labels in test_ds.take(1):
    for i in range(9):
        ax = plt.subplot(3, 3, i + 1)
        plt.imshow(np.array(images[i]).astype("uint8"))
        plt.title(int(labels[i]))
        plt.axis("off")
```

```python
# Loading data from custom module and defining image size
import data_loader

image_size = (224, 224)
train_ds, val_ds = data_loader.load_data()
```

### MobileNetV2 Model

```python
# Loading MobileNetV2 model
model = keras.models.load_model('MobileNetV2_model.keras')
```

```python
# Making prediction on validation dataset
pred = model.predict(test_ds)
```

```python
# Viewing prediction logit values
pred
```

The above shows an array of logit values (real numbers that range from negative infinity to infinity). Positive logit values indicate a probability greater than 0.5 that the instance is in class 1 (forest). Negative logit values indicate a probability less than 0.5, so the model would predict that the instance is in class 0 (buildings).

```python
# Comparing logit value and class for first 9 images
plt.figure(figsize=(10, 10))
for images, labels in test_ds.take(1):
    for i in range(9):
        ax = plt.subplot(3, 3, i + 1)
        plt.imshow(np.array(images[i]).astype("uint8"))
        plt.title(f'Logit:{pred.ravel()[i]:.2f},  Actual:{int(labels[i])}')
        plt.axis("off")
```

```python
# Retrieving images labels
image_labels = np.concatenate([y for x, y in test_ds], axis=0)
image_labels
```

```python
# Calculating TPR, FPR, and AUC for ROC curve
from sklearn import metrics

fpr, tpr, thresholds = metrics.roc_curve(image_labels, pred)
auc = metrics.auc(fpr, tpr)
```

```python
# Plotting ROC curve for MobileNetV2 model
plt.plot(fpr,tpr,color='darkorange',label=f'ROC curve (area = {auc})')
plt.legend(loc="lower right")
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')

plt.show()
```
